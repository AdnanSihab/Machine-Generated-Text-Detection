```python
import pandas as pd
import csv
import json
```

```python
lst = []
data_csv = open("/data_csv.txt", "a")


f = open('/kaggle/input/subtaska/subtaskA_dev_monolingual.jsonl',"r")
a = f.readlines()
print(len(a))
```

```python
for i in range(5000):
    data1 = json.loads(a[i])
    temp = (data1["text"],data1["label"])
    lst.append(temp)
```

```python
csv_file_name = "/kaggle/working/output_val.csv"


with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)


    csv_writer.writerow(['text', 'label'])


    csv_writer.writerows(lst)

print(f'Data has been written to {csv_file_name}.')
```
